# Phần I: Giới thiệu & Tổng quan về Unsupervised Learning

---
## 1. LỜI MỞ ĐẦU & DẪN DẮT VẤN ĐỀ

### Vấn đề thực tế: Big Data Era

> **"Mỗi ngày, chúng ta tạo ra khoảng 2.5 quintillion byte dữ liệu"**  
> *(1 quintillion = 1,000,000,000,000,000,000)*

**Nguồn dữ liệu:**
- Mạng xã hội: 500 triệu tweets/ngày
- Video: 720,000 giờ video được upload lên YouTube mỗi ngày
- Giao dịch: Hàng tỷ transaction từ thương mại điện tử
- Y tế: Hình ảnh y khoa, dữ liệu bệnh án
- Khoa học: Dữ liệu từ các thí nghiệm, cảm biến IoT

### Thách thức

**Phần lớn dữ liệu này ở dạng "thô" (Raw Data):**
- Chưa được phân loại
- Không có nhãn (labels)
- Không có chú thích (annotations)

### Câu hỏi đặt ra

**Làm thế nào để máy tính có thể tự động trích xuất thông tin hữu ích từ mớ dữ liệu hỗn độn đó mà không cần con người phải ngồi chỉ dạy từng chút một?**

### Giải pháp

**→ Unsupervised Learning (Học không giám sát)**

---
## 2. ĐỊNH NGHĨA CỐT LÕI: UNSUPERVISED LEARNING LÀ GÌ?

### Khái niệm

**Unsupervised Learning** là một nhánh của Machine Learning (Học máy), trong đó các thuật toán được huấn luyện trên các tập dữ liệu **không có nhãn** (Unlabeled data).

### Đặc điểm chính

1. **Không có Ground Truth**
   - Thuật toán không biết kết quả đầu ra sẽ trông như thế nào

2. **Mục tiêu: khám phá (Exploration), không phải dự đoán (Prediction)**
   - Tìm kiếm **cấu trúc ẩn** (hidden structures)
   - Phát hiện **mẫu** (patterns)
   - Nhận diện **điểm tương đồng** trong dữ liệu

3. **Vai trò của máy học**
   - Đóng vai trò như một **explorer**
   - Tự động học và tổ chức thông tin

### Công thức toán học

**Dữ liệu đầu vào:** Chỉ có $X = \{x_1, x_2, ..., x_n\}$

Trong đó:
- $x_i \in \mathbb{R}^d$: mỗi điểm dữ liệu có $d$ đặc trưng (features)
- **KHÔNG có** $y$ (nhãn/label)

**Mục tiêu:** Tìm hàm $f$ sao cho $f(X) \rightarrow$ cấu trúc/pattern có ý nghĩa


---
## 3. PHÂN BIỆT: SUPERVISED và UNSUPERVISED LEARNING

### So sánh qua ví dụ

| Tiêu chí | **Supervised Learning** | **Unsupervised Learning** |
|----------|------------------------|---------------------------|
| **Dữ liệu** | $(X, y)$ - có input VÀ output | Chỉ có $X$ - chỉ có input |
| **Mục tiêu** | **Dự đoán** (Prediction) | **Khám phá** (Discovery) |
| **Câu hỏi** | "Đây là chó hay mèo?" | "Có bao nhiêu nhóm động vật?" |
| **Ví dụ bài toán** | Classification, Regression | Clustering, Anomaly Detection |
| **Đánh giá** | Accuracy, F1-score (có ground truth) | Silhouette score, Inertia (không có ground truth) |

---
## 4. CÁC BÀI TOÁN CHÍNH CỦA UNSUPERVISED LEARNING

### Bức tranh tổng quan

Unsupervised Learning giải quyết **3 bài toán cốt lõi:**

### A. CLUSTERING (Phân cụm)

#### Định nghĩa
Quá trình **nhóm các điểm dữ liệu** lại với nhau sao cho:
- Các điểm **trong cùng một cụm** (cluster) có tính chất **giống nhau nhất**
- Các điểm **ở các cụm khác nhau** có tính chất **khác biệt tối đa**

#### Ứng dụng thực tế
- **Marketing:** Phân khúc khách hàng (Customer Segmentation) để cá nhân hóa quảng cáo
- **Y tế:** Nhóm bệnh nhân có triệu chứng tương tự
- **Sinh học:** Phân loại gen, protein
- **Thương mại điện tử:** Gom nhóm sản phẩm tương tự

#### Thuật toán phổ biến
- K-Means
- DBSCAN
- Hierarchical Clustering
- Gaussian Mixture Models (GMM)

### B. ASSOCIATION RULE / PATTERN DISCOVERY (Khám phá luật kết hợp)

#### Định nghĩa
Tìm ra các **quy luật**, **sự kiện** thường xuyên xuất hiện **cùng nhau** trong cơ sở dữ liệu lớn.

#### Ứng dụng thực tế
- **Retail:** Market Basket Analysis (Phân tích giỏ hàng)
- **E-commerce:** Hệ thống gợi ý sản phẩm
- **Web Mining:** Phân tích hành vi người dùng

#### Ví dụ kinh điển
**"Beer & Diapers" Case:**
- Walmart phát hiện: Khách hàng mua tã trẻ em thường mua bia
- Lý do: Bố trẻ đi mua tã cho con, mua luôn bia cho mình
- Hành động: Đặt bia gần tã → Tăng doanh số!

#### Thuật toán phổ biến
- Apriori
- FP-Growth
- ECLAT

### C. ANOMALY DETECTION (Phát hiện bất thường)

#### Định nghĩa
Xác định các **điểm dữ liệu**, **sự kiện** hoặc **quan sát hiếm gặp**, đi lệch khỏi hành vi bình thường của toàn bộ tập dữ liệu.

#### Ứng dụng thực tế
- **Tài chính:** Phát hiện gian lận thẻ tín dụng (Credit Card Fraud Detection)
- **Cybersecurity:** Phát hiện xâm nhập mạng (Network Intrusion Detection)
- **Sản xuất:** Phát hiện lỗi hỏng hóc máy móc
- **Y tế:** Phát hiện bệnh hiếm gặp hoặc đột biến gen

#### Thuật toán phổ biến
- Isolation Forest
- Local Outlier Factor (LOF)
- One-Class SVM
- Autoencoders

# Phần II: Một số kỹ thuật phân cụm cơ bản. Phương pháp phân cụm K-Means

# Clustering (Phân cụm dữ liệu)

## 1. Giới thiệu vấn đề
**Cluster Analysis (phân tích phân cụm)** là quá trình chia tập dữ liệu thành nhiều nhóm sao cho các phần tử trong cùng một nhóm có độ tương đồng cao, trong khi các phần tử thuộc các nhóm khác nhau có độ khác biệt lớn.

Ví dụ, trong một hệ thống thương mại điện tử, dữ liệu khách hàng có thể được phân cụm dựa trên hành vi mua sắm như:
- nhóm khách hàng thường xuyên mua hàng
- nhóm khách hàng mua hàng theo mùa
- nhóm khách hàng ít hoạt động

Việc phân cụm giúp doanh nghiệp hiểu rõ hơn về hành vi người dùng và xây dựng các chiến lược tiếp thị phù hợp.

Giả sử ta có một tập dữ liệu và ta cần phải nhóm các dữ liệu có tính chất tương tự nhau vào các cụm khác nhau chưa biết trước. Một cách đơn giản để mô phỏng bài toán này là biểu diễn bằng hình học. Các dữ liệu có thể coi là các điểm trong không gian và khoảng cách giữa các điểm có thể được coi là thông số mức độ giống nhau của chúng. 2 điểm càng gần nhau thì chúng càng giống nhau.

## 2. Các thuật toán phân cụm phổ biến
Hiện nay có nhiều thuật toán phân cụm khác nhau được phát triển trong lĩnh vực Machine Learning. Mỗi thuật toán có ưu điểm và hạn chế riêng, phù hợp với các loại dữ liệu và mục đích khác nhau. Một số thuật toán phổ biến bao gồm:
- **K-Means:** Đây là một trong những thuật toán phân cụm cơ bản và được sử dụng rộng rãi nhất. Thuật toán chia tập dữ liệu thành một số cụm xác định trước và biểu diễn mỗi cụm bằng một điểm trung tâm.
- **DBSCAN:** Là thuật toán phân cụm dựa trên mật độ dữ liệu. Phương pháp này có khả năng phát hiện các cụm có hình dạng phức tạp và xử lý tốt các điểm nhiễu trong dữ liệu.
- **Hierarchical Clustering:** Thuật toán này xây dựng cấu trúc phân cấp giữa các cụm dữ liệu dưới dạng một cây phân cấp, cho phép quan sát mối quan hệ giữa các cụm ở nhiều mức độ khác nhau.

Các thuật toán trên đại diện cho những hướng tiếp cận phổ biến trong phân cụm dữ liệu, bao gồm **phân cụm dựa trên centroid**, **phân cụm dựa trên mật độ** và **phân cụm phân cấp**. Trong các phần tiếp theo, báo cáo sẽ trình bày chi tiết hơn về nguyên lý và cách hoạt động của từng thuật toán.

# K-Means

## 1. Ý tưởng chính của K-Means
Thuật toán chia tập dữ liệu thành **K cụm (clusters)** sao cho:
- Các điểm trong cùng 1 cụm thì giống nhau.
- Các điểm khác cụm thì khác nhau.

Mỗi cụm được đại diện bởi 1 điểm trung tâm gọi là **centroid**.

Mục tiêu của K-means là **tối thiểu hóa** tổng khoảng cách từ các điểm dữ liệu đến **centroid** của cụm.

## 2. Quy trình hoạt động của K-Means
Giả sử ta chia dữ liệu thành **K cụm**.

**Bước 1: Khởi tạo centroid**

Chọn ngẫu nhiên **K điểm** trong dữ liệu làm centroid ban đầu.

**Bước 2: Gán điểm dữ liệu vào cụm gần nhất**

Với mỗi điểm dữ liệu:
- Tính khoảng cách tới từng centroid.
- Gán điểm đó vào cluster có centroid gần nhất.

**Hàm tính khoảng cách Euclid (Euclidean distance):**

<p align="center">
  $d(x,y) = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$
</p>

**Bước 3: Cập nhật centroid**

Sau khi gán các điểm vào cluster:
- Tính trung bình của tất cả điểm trong cluster
- Trung bình này trở thành centroid mới

**Ví dụ.** Cluster A có các điểm:
$(1,2) - (2,3) - (3,4)$

Centroid mới: $(\frac{1+2+3}{3} , \frac{2+3+4}{3}) = (2 , 3)$

**Bước 4: Lặp lại**

Lặp lại Bước 2 và 3 cho đến khi **centroid không thay đổi** nữa hoặc đạt **số lần lặp tối đa**.

````python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

class KMeans():
  def __init__(self, k, max_iteration=10):
      self.k = k
      self.max_iteration = max_iteration
      self.all_centroids = []
      self.all_labels = []

  # Hàm thuật toán K-Means lấy đầu vào là một bộ dữ liệu và số lượng cluster K. Trả về tâm của K cụm
  def fit(self, dataSet):
      # Khởi tạo ngẫu nhiên k centroids
      numFeatures = dataSet.shape[1]
      centroids = self.get_random_centroids(numFeatures, self.k)
      self.all_centroids.append(centroids)
      self.all_labels.append(None)

      # Khởi tạo các biến iterations, oldCentroids
      iterations = 0
      oldCentroids = None

      # Vòng lặp cập nhật centroids trong thuật toán k-Means
      while not self.should_stop(oldCentroids, centroids, iterations):
          # Lưu lại centroids cũ cho quá trình kiểm tra hội tụ
          oldCentroids = centroids
          iterations += 1

          # Gán nhãn cho mỗi diểm dữ liệu dựa vào centroids
          labels = self.get_labels(dataSet, centroids)
          self.all_labels.append(labels)

          # Cập nhật centroids dựa vào nhãn dữ liệu
          # print('Old centroids: ', centroids)
          centroids = self.get_centroids(dataSet, labels, self.k)
          # print('new centroids: ', centroids)
          self.all_centroids.append(centroids)

      return centroids

  # Hàm khởi tạo centroids ngẫu nhiên
  def get_random_centroids(self, numFeatures, k):
    return np.random.rand(k, numFeatures)
    # return np.array([[-5., -5.],
    #                  [4., 6.]])

  # Hàm này trả về nhãn cho mỗi điểm dữ liệu trong datasets
  def get_labels(self, dataSet, centroids):
      # Với mỗi quan sát trong dataset, lựa chọn centroids gần nhất để gán label cho dữ liệu.
      labels = []
      for x in dataSet:
        # Tính khoảng cách tới các centroids và cập nhận nhãn
        distances = np.sum((x-centroids)**2, axis=1)
        label = np.argmin(distances)
        labels.append(label)
      return labels

  # Hàm này trả về True hoặc False nếu k-Means hoàn thành. Điều kiện k-Means hoàn thành là
  # thuật toán vượt ngưỡng số lượng vòng lặp hoặc centroids ngừng thay đổi
  def should_stop(self, oldCentroids, centroids, iterations):
      if iterations > self.max_iteration:
        return True
      return np.all(oldCentroids == centroids)

  # Trả về tọa độ mới cho k centroids của mỗi chiều.
  def get_centroids(self, dataSet, labels, k):
      centroids = []
      for j in np.arange(k):
        # Lấy index cho mỗi centroids
        idx_j = np.where(np.array(labels) == j)[0]
        centroid_j = dataSet[idx_j, :].mean(axis=0)
        centroids.append(centroid_j)
      return np.array(centroids)

# cluster_std: độ lệch chuẩn của tập dữ liệu (càng nhỏ thì các điểm nằm càng gần nhau)
dataset, _ = make_blobs(n_samples=250, cluster_std=3.0, random_state=123)

kmean = KMeans(k=2, max_iteration=8)
centroids = kmean.fit(dataset)

gs = GridSpec(nrows=3, ncols=3)
plt.figure(figsize = (15, 15))
plt.subplots_adjust(wspace=0.2,hspace=0.4)
colors = ['green', 'blue']
labels = ['cluster 1', 'cluster 2']

for i in np.arange(len(kmean.all_centroids)):
  ax = plt.subplot(gs[i])
  if i == 0:
    centroids_i = kmean.all_centroids[i]
    plt.scatter(dataset[:, 0], dataset[:, 1], s=50, alpha=0.5, color='red')
    for j in np.arange(kmean.k):
      plt.scatter(centroids_i[j, 0], centroids_i[j, 1], marker='x', s=100, color='red')
    plt.title('All points in original dataset')
  else:
    # Lấy centroids và labels tại bước thứ i
    centroids_i = kmean.all_centroids[i]
    labels_i = kmean.all_labels[i]
    # Visualize các điểm cho từng cụm
    for j in np.arange(kmean.k):
      idx_j = np.where(np.array(labels_i) == j)[0]
      plt.scatter(dataset[idx_j, 0], dataset[idx_j, 1], color=colors[j], label=labels[j], s=50, alpha=0.3, lw = 0)
      plt.scatter(centroids_i[j, 0], centroids_i[j, 1], marker='x', color=colors[j], s=100, label=labels[j])
    plt.title(r'iteration {}'.format(i))
````

## 3. Hàm mục tiêu của thuật toán K-Means:

**WCSS (Within-Cluster Sum of Squares)** - tổng bình phương khoảng cách của các điểm tới centroid của cụm.

<p align="center">
  $$J = \sum_{i=1}^{K} \sum_{x \in C_i} \|x - \mu_i\|^2$$
</p>

**Trong đó:**
- $K$ : số cụm (clusters)
- $C_i$ : tập các điểm dữ liệu thuộc cụm thứ $i$
- $x$ : một điểm dữ liệu trong cụm
- $\mu_i$ : centroid (trung tâm) của cụm $i$
- $\|x - \mu_i\|^2$ : bình phương khoảng cách Euclidean giữa điểm $x$ và centroid của cụm thứ $i$

**Ý nghĩa:**

Hàm mục tiêu giúp thuật toán đánh giá chất lượng của việc phân cụm.
- Nếu giá trị hàm mục tiêu nhỏ $\to$ các điểm dữ liệu nằm gần centroid $\to$ cụm chặt và tốt.
- Nếu giá trị hàm mục tiêu lớn $\to$ các điểm phân tán $\to$ cụm kém chất lượng.

Vì vậy, trong quá trình hoạt động, thuật toán K-means sẽ liên tục cập nhật centroid và phân cụm lại dữ liệu để giảm giá trị của hàm mục tiêu để đạt được giá trị tối thiểu.

## 4. Hạn chế và giải pháp

### **Cần biết trước số cụm**
Dễ thấy rằng điều kiện đầu vào của giải thuật bắt buộc phải chỉ rõ giá trị của K, nhưng trong thực tế không phải lúc nào ta cũng có thể biết trước được có bao nhiêu cụm. Vấn đề này có thể cải thiện dựa vào một số phương pháp sau:
- Phương pháp Elbow *(phổ biến nhất)*.
- Phương pháp X-means.
## **Phương pháp Elbow**
Ý tưởng chính của phương pháp này là quan sát sự thay đổi của WCSS khi số lượng cụm $K$ tăng dần. WCSS đo tổng bình phương khoảng cách giữa các điểm dữ liệu và centroid của cụm mà chúng thuộc về. Khi $K$ tăng, giá trị WCSS thường giảm do các điểm dữ liệu nằm gần centroid hơn.

Phương pháp Elbow được thực hiện theo các bước sau:
- **Bước 1:** Chạy thuật toán K-Means với nhiều giá trị $K$ khác nhau (ví dụ từ 1 đến 10).
- **Bước 2:** Tính giá trị WCSS cho từng giá trị $K$.
- **Bước 3:** Vẽ đồ thị biểu diễn mối quan hệ giữa $K$ và WCSS.
- **Bước 4:** Xác định điểm “khuỷu tay” (elbow) trên đồ thị, tại đó tốc độ giảm của WCSS bắt đầu chậm lại.

Giá trị $K$ tại điểm khuỷu tay thường được xem là số cụm phù hợp cho tập dữ liệu.

````python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


dataset, _ = make_blobs(n_samples=250, cluster_std=3.0, random_state=123)
losses = []
K = 10
for i in range(1, K):
  # 1.  Huấn luyện với số cụm = i
  kmeans_i = KMeans(n_clusters=i, random_state=0)
  kmeans_i.fit(dataset)
  # 2. Tính _hàm biến dạng_
  # 2.1. Khoảng cách tới toàn bộ centroids
  d2centroids = cdist(dataset, kmeans_i.cluster_centers_, 'euclidean') # shape (n, k)
  # 2.2. Khoảng cách tới centroid gần nhất
  min_distance = np.min(d2centroids, axis=1) # shape (n)
  loss = np.sum(min_distance)
  losses.append(loss)

plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, K), losses, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
````

Trong hình trên thì ta thấy vị trí của điểm khuỷu tay chính là $k = 2$ vì khi số lượng cụm lớn hơn $2$ thì tốc độ suy giảm của *hàm biến dạng* dường như không đáng kể so với trước đó.

***
### **Khởi tạo ảnh hưởng tới chất lượng**
Vị trí của các centroid được khởi tạo có ý nghĩa lớn tới tốc độ và độ chính xác của thuật toán. Nếu các centroid ban đầu được chọn **quá gần nhau**, nhiều centroid có thể nằm trong cùng một vùng dữ liệu. Khi đó, một số cụm sẽ bị bỏ trống hoặc phân bố dữ liệu không hợp lý. Điều này khiến thuật toán cần nhiều vòng lặp hơn để điều chỉnh lại vị trí centroid, thậm chí có thể dẫn đến kết quả phân cụm không tối ưu. Để phần nào khắc phục được tình trạng này, người ta có thể sử dụng một số cách như:
- Chạy giải thuật nhiều lần với mỗi cách khởi tạo khác nhau và chọn lấy cách đưa ra kết quả tốt nhất.
- Sử dụng phương pháp khởi tạo K-Means++.
## **K-Means++**
Trong K-Means truyền thống, các centroid thường được khởi tạo ngẫu nhiên. Tuy nhiên, cách khởi tạo này có thể dẫn đến việc các centroid nằm quá gần nhau, làm giảm chất lượng phân cụm hoặc khiến thuật toán hội tụ chậm.

Phương pháp K-Means++ được đề xuất nhằm lựa chọn các centroid ban đầu một cách hợp lý hơn. Ý tưởng của phương pháp là chọn các centroid sao cho chúng cách xa nhau nhất có thể. Quy trình khởi tạo được thực hiện như sau:
- **Bước 1:** Chọn ngẫu nhiên một điểm dữ liệu trong tập dữ liệu làm centroid đầu tiên.
- **Bước 2:** Với mỗi điểm dữ liệu còn lại, tính khoảng cách từ điểm đó đến centroid gần nhất đã được chọn.
- **Bước 3:** Chọn centroid tiếp theo với xác suất tỷ lệ với bình phương khoảng cách đã tính.
- **Bước 4:** Lặp lại quá trình cho đến khi đủ $K$ centroid.

Nhờ cơ chế lựa chọn này, các centroid ban đầu thường phân bố tốt hơn trong không gian dữ liệu, từ đó giúp **tăng tốc độ hội tụ và cải thiện chất lượng phân cụm**.

````python
import numpy as np
import matplotlib.pyplot as plt
import math

mean_01 = np.array([0.0, 0.0])
cov_01 = np.array([[1, 0.3], [0.3, 1]])
dist_01 = np.random.multivariate_normal(mean_01, cov_01, 100)

mean_02 = np.array([6.0, 7.0])
cov_02 = np.array([[1.5, 0.3], [0.3, 1]])
dist_02 = np.random.multivariate_normal(mean_02, cov_02, 100)

mean_03 = np.array([7.0, -5.0])
dist_03 = np.random.multivariate_normal(mean_03, cov_01, 100)

mean_04 = np.array([2.0, -7.0])
cov_04 = np.array([[1.2, 0.5], [0.5, 1.3]])
dist_04 = np.random.multivariate_normal(mean_04, cov_01, 100)

data = np.vstack((dist_01, dist_02, dist_03, dist_04))
np.random.shuffle(data)

def distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))


def initialize(data, k):
    centroids = []
    centroids.append(data[np.random.randint(data.shape[0])])

    all_centroids = [np.array(centroids)]

    for _ in range(k - 1):
        distances = []
        for point in data:
            min_dist = min([distance(point, c) for c in centroids])
            distances.append(min_dist)

        next_centroid = data[np.argmax(distances)]
        centroids.append(next_centroid)

        all_centroids.append(np.array(centroids))

    return all_centroids


def plot_all(data, all_centroids):

    n = len(all_centroids)
    cols = 2
    rows = math.ceil(n / cols)

    fig, axes = plt.subplots(rows, cols, figsize=(12, 5 * rows))
    axes = axes.flatten()

    for i, centroids in enumerate(all_centroids):
        ax = axes[i]

        ax.scatter(data[:, 0], data[:, 1], marker='.', color='gray', label='Data Points')

        if centroids.shape[0] > 1:
            ax.scatter(centroids[:-1, 0], centroids[:-1, 1], color='black', label='Selected Centroids')

        ax.scatter(centroids[-1, 0], centroids[-1, 1], color='red', label='Next Centroid')

        ax.set_title(f'Select {centroids.shape[0]}th Centroid')
        ax.set_xlim(-5, 12)
        ax.set_ylim(-10, 15)
        ax.legend()

    plt.tight_layout()
    plt.show()


# Run
all_centroids = initialize(data, k=4)
plot_all(data, all_centroids)
````

# Phần III: Một số kỹ thuật phân cụm nâng cao, Phương pháp phân cụm DBSCAN & Hierachical

## DBSCAN

### Một số định nghĩa
#### 1. Vùng lân cận epsilon (Eps-neighborhood)

Vùng lân cận epsilon (Eps-neighborhood) của một điểm dữ liệu được định nghĩa là tập hợp tất cả các điểm dữ liệu $P$ nằm trong phạm vi bán kính epsilon (kí hiệu $\epsilon$) xung quanh điểm $P$. Tập hợp này được kí hiệu là:

$N_\epsilon(P) = \{ Q \in \mathcal{D} : d(P<Q) \leq \epsilon \}$, trong đó:

- $\mathcal{D}$ là tập hợp tất cả các điểm dữ liệu sử dụng trong tập huấn luyện;
- $d(P,Q)$ có thể là là khoảng cách giữa hai điểm p và q (thường là khoảng cách Euclidean)

$\to$ Trong dữ liệu 2 chiều, ta vẽ một hình tròn tâm $P$, bán kính $\epsilon$. Các điểm nằm trên biên hoặc bên trong hình tròn được xác định là **vùng lân cận của điểm $P$**.

#### 2. Khả năng tiếp cận trực tiếp mật độ (directly density-reachable)
Khả năng tiếp cận trực tiếp mật độ (directly density-reachable) dề cập tới việc một điểm có thể **tiếp cận trực tiếp** tới một điểm dữ liệu khác. Theo đó, một điểm $Q$ được coi là khả tiếp cận trực tiếp điểm $P$ tương ứng với bộ tham số `epsilon`, `minPts`:

- $Q$ nằm trong vùng lân cận $\epsilon$ của $P$: $Q \in N_\epsilon(P)$
- Số lượng các điểm dữ liệu nằm trong vùng lân cận $\epsilon$ tối thiểu **minPts**: $|N_\epsilon(P)| \geq minPts$

$\to$ Như vậy một điểm dữ liệu có thể tiếp cận được trực tiếp tới một điểm khác phải đáp ứng được cả hai yếu tố về ***khoảng cách giữa chúng*** và ***mật độ các điểm dữ liệu*** trong vùng lân cận `epsilon` phải tối thiểu bằng `minPts`

#### 3. Khả năng tiếp cận mật độ (density-reachable)
Khả năng tiếp cận mật độ (density-reachable) liên quan đến cách hình thành một chuỗi liên kết điểm trong cụm. Cụ thể là trong một tập hợp chuỗi điểm $\{ P_i\}_{i=1}^n \subset \mathcal{D}$ mà nếu như bất kì một điểm $P_i$ nào cũng đều có thể tiếp cận trực tiếp mật độ bởi $P_i$ theo tham số `epsilon` và `minPts` thì khi đó ta nói điểm $P_n$ có khả năng kết nối mật độ tới điểm
$P_i$.

$\to$ 2 điểm $P_i$ và $P_j$ bất kì thuộc chuỗi $\{ P_i\}_{i=1}^n$ thoả mãn $i < j$ thì đều có khả năng kết nối mật độ tới $P_i$.

Như vậy, 2 điểm bất kì có khả năng kết nối mật độ với nhau thì sẽ thuộc cùng một cụm và các điểm trong chuỗi $\{ P_i\}_{i=1}^n$ đều được phân về cùng cụm. Khả năng tiếp cận mật độ thể hiện sự mở rộng phạm vi của một cụm dữ liệu dựa trên liên kết theo chuỗi. Xuất phát từ một điểm dữ liệu ta có thể tìm được các điểm có khả năng kết nối mật độ tới nó theo lan truyền chuỗi để xác định cụm.

#### Phân loại dạng điểm trong DBSCAN
<p align="center">
  <img src="https://github.com/newfrogg/DataViZ/blob/main/images/DBSCAN_types-of-points.png?raw=1" width="500" />
</p>

- Điểm lõi (core): trung tâm của cụm
- Điểm biên (border): điểm nằm trong cụm nhưng không phải là tâm
- Điểm nhiễu (noise, outliner): điểm nằm ngoài tất cả các cụm

Tóm tắt:
| Khái niệm               | Điều kiện liên quan đến vùng lân cận                              | Ý nghĩa                                                                 |
|-------------------------|-------------------------------------------------------------------|-------------------------------------------------------------------------|
| **Core point**          | $|N_\epsilon(P)| \geq minPts$ (bao gồm chính $P$)                              | Điểm nằm ở **trung tâm** vùng mật độ cao → là "xương sống" của cụm     |
| **Border point**        | $|N_\epsilon(P)| < minPts$, nhưng $P$ nằm trong $N_\epsilon(P)$ của ít nhất 1 core point | Điểm nằm ở **rìa** cụm                                                   |
| **Noise / Outlier**     | Không thỏa mãn hai điều trên (không thuộc vùng lân cận của core point nào) | Điểm nhiễu, cô lập                                                     |
| **Directly density-reachable** | $Q \in N_\epsilon(P)$ và $P$ là core point                                      | q có thể được "kéo" trực tiếp vào cụm từ p                              |
| **Density-reachable**   | Có chuỗi các core point nối tiếp nhau để từ $P$ đến $Q$              | $Q$ thuộc cùng cụm với $P$ (qua đường truyền mật độ)                        |



**DBSCAN** (Density-Based Spatial Clustering of Applications with Noise) là một trong những thuật toán **phân cụm** (clustering) phổ biến và mạnh mẽ nhất trong học máy không giám sát. Nó đặc biệt phù hợp khi dữ liệu có các cụm hình dạng bất kỳ và chứa **nhiễu** (outliers).

### Ý tưởng
DBSCAN coi một **cụm** là một vùng có **mật độ cao** (nhiều điểm dữ liệu gần nhau), được ngăn cách bởi các vùng có **mật độ thấp**.  
Các điểm nằm ở vùng mật độ thấp → được coi là **nhiễu**.

Khác với K-Means (giả định cụm hình cầu, cần biết trước số cụm), DBSCAN:
- Không cần chỉ định số cụm trước
- Tự động phát hiện nhiễu
- Tìm được cụm hình dạng tùy ý (dài, cong, vòng, lồng nhau…)

### Hai tham số quan trọng nhất

| Tham số     | Ký hiệu | Ý nghĩa                                                                 | Giá trị điển hình                  |
|-------------|---------|--------------------------------------------------------------------------|-------------------------------------|
| Bán kính    | `epsilon` | Khoảng cách tối đa để hai điểm được coi là hàng xóm (**neighbors**)     | Phụ thuộc dữ liệu, thường 0.1–1.0 (sau khi scale) |
| Số điểm tối thiểu | `MinPts`  | Số điểm tối thiểu (bao gồm chính nó) trong bán kính `epsilon` để xác định 1 điểm lõi (**core point**) | Thường 4–10 (giá trị mặc định phổ biến là 5) |

### Phân loại các điểm dữ liệu trong DBSCAN

Mỗi điểm dữ liệu được phân loại thành 3 loại:

| Loại điểm       | Định nghĩa                                                                 | Đặc điểm                                                                 |
|-----------------|----------------------------------------------------------------------------|---------------------------------------------------------------------------|
| **Core point**  | Có nhiều hơn hoặc bằng `MinPts` điểm (bao gồm chính nó) nằm trong bán kính `epsilon`               | Nằm ở **trung tâm** của cụm, mật độ cao                                   |
| **Border point**| Không phải core point, nhưng nằm trong bán kính ε của **ít nhất 1 core point** | Nằm ở **rìa** cụm                                                        |
| **Noise / Outlier** | Không phải core point lẫn border point (không thuộc bất kỳ cụm nào)       | Điểm nhiễu, cô lập                                                       |

### Thuật toán DBSCAN

1. Chọn một điểm $P$ bất kỳ chưa được thăm.
2. Nếu $P$ là **core point** $\to$ tạo một cụm mới, thêm tất cả các điểm **trực tiếp reachable** và **density-reachable** vào cụm.

3. Mở rộng cụm bằng cách:
   - Tìm tất cả điểm trong `epsilon` của $P$
   - Nếu gặp core point khác $\to$ tiếp tục mở rộng từ điểm đó (giống BFS)
4. Nếu $P$ không phải core point $\to$ đánh dấu là **noise** (tạm thời)

5. Lặp lại với các điểm chưa thăm cho đến khi hết dữ liệu.
6. Các điểm noise có thể được gán lại thành border point nếu sau này thuộc cụm khác.

### Ưu điểm nổi bật

- Tìm được **cụm hình dạng bất kỳ** (không giới hạn hình cầu, elip)
- Tự động phát hiện và loại bỏ **nhiễu/outlier**
- Không cần biết trước số lượng cụm
- Hoạt động tốt với dữ liệu có mật độ không đồng đều

### Nhược điểm & Thách thức thực tế

| Nhược điểm                          | Giải pháp dùng                                                                 |
|-------------------------------------|---------------------------------------------------------------------------------------------|
| Rất nhạy với việc chọn `epsilon` và `MinPts` | Dùng **k-distance plot** (thường k = MinPts-1) để chọn `epsilon` hợp lý                              |
| Không xử lý tốt khi các cụm có **mật độ rất khác nhau** | Cân nhắc dùng HDBSCAN (phiên bản cải tiến) hoặc OPTICS                                       |
| Hiệu suất kém trên dữ liệu chiều cao (>20–30 chiều) | Giảm chiều trước (PCA, UMAP, t-SNE) hoặc dùng phiên bản tăng tốc (HDBSCAN, RAPIDS cuML)     |
| Không hoạt động tốt với dữ liệu rất thưa (sparse) | Chuẩn hóa dữ liệu (StandardScaler, MinMaxScaler, đặc biệt với khoảng cách Euclidean)   |

### Cách chọn tham số ε và MinPts (mẹo thực tế)

1. Chuẩn hóa dữ liệu về cùng thang đo (rất quan trọng!)
2. Vẽ **k-distance plot**:
   - Tính khoảng cách đến điểm hàng xóm thứ ($k \approx$ `MinPts`-1) cho mọi điểm
   - Sắp xếp tăng dần $\to$ tìm “điểm khuỷu tay” (elbow) $\to$ chọn `epsilon` gần giá trị đó
3. `MinPts` thường chọn:
   - 2 $\times$ số chiều (dimensionality)
   - Giá trị lớn hơn thì xác định ít cụm hơn, ít nhiễu hơn nhưng có thể bỏ sót một số cụm nhỏ hơn.

### So sánh DBSCAN với các thuật toán phổ biến khác

| Thuật toán   | Cần biết số cụm trước? | Hình dạng cụm       | Xử lý nhiễu tốt? | Mật độ khác nhau | Phức tạp thời gian     |
|--------------|------------------------|---------------------|-------------------|------------------|------------------------|
| K-Means      | Có                     | Hình cầu            | Kém               | Kém              | O(n)                   |
| Hierarchical | Không (cắt dendrogram) | Tùy thuộc linkage   | Trung bình        | Trung bình       | O(n²) – O(n³)          |
| **DBSCAN**   | **Không**              | **Bất kỳ**          | **Rất tốt**       | Trung bình       | O(n log n) với index   |
| HDBSCAN      | Không                  | Bất kỳ              | Rất tốt           | **Tốt**          | O(n log n)             |

### Tổng hợp

**DBSCAN**
- phân cụm dựa trên **mật độ**  
- Các core points nối nhau $\to$ tạo cụm  
- Điểm có mặt độ thấp được coi là nhiễu hoặc ngoại lai.
- Phụ thuộc vào 2 tham số chính: `MinPts`, `epsilon`  
$\to$ Đặc trung cho các tác vụ phân cụm có hình dạng lạ, mà không biết số cụm và lọc nhiễu.

````python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import MinMaxScaler
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import seaborn as sns
import numpy as np

data = pd.read_csv("https://raw.githubusercontent.com/newfrogg/DataViZ/refs/heads/main/dataset/shopping-data.csv", header=0, index_col=0)
print(data.shape)
data.head()

# Lấy ra thu nhập va điểm shopping
X = data.iloc[:, 2:4].values
print(X.shape)
std = MinMaxScaler()
X_std = std.fit_transform(X)
````

````python
from sklearn.neighbors import NearestNeighbors


# Xây dựng mô hình k-Means với k=10
neighbors = 10
nbrs = NearestNeighbors(n_neighbors=neighbors ).fit(X_std)

# Ma trận khoảng cách distances: (N, k)
distances, indices = nbrs.kneighbors(X_std)

# Lấy ra khoảng cách xa nhất từ phạm vi láng giềng của mỗi điểm và sắp xếp theo thứ tự giảm dần.
distance_desc = sorted(distances[:, neighbors-1], reverse=True)

# Vẽ biểu đồ khoảng cách xa nhất ở trên theo thứ tự giảm dần
plt.figure(figsize=(12, 8))
plt.plot(list(range(1,len(distance_desc )+1)), distance_desc)
plt.axhline(y=0.12)
plt.text(2, 0.12, 'y = 0.12', fontsize=12)
plt.axhline(y=0.16)
plt.text(2, 0.16, 'y = 0.16', fontsize=12)
plt.ylabel('distance')
plt.xlabel('indice')
plt.title('Sorting Maximum Distance in k Nearest Neighbor of kNN')
````

````python
from matplotlib.gridspec import GridSpec
import warnings
warnings.simplefilter("ignore", category=RuntimeWarning)

def _plot_kmean_scatter(X, labels, gs, thres):
    '''
    X: dữ liệu đầu vào
    labels: nhãn dự báo
    '''
    # lựa chọn màu sắc
    num_classes = len(np.unique(labels))
    palette = np.array(sns.color_palette("hls", num_classes))

    # vẽ biểu đồ scatter
    ax = plt.subplot(gs)
    sc = ax.scatter(X[:,0], X[:,1], lw=0, s=40, c=palette[labels.astype(int)])

    # thêm nhãn cho mỗi cluster
    txts = []

    for i in range(num_classes):
        # Vẽ text tên cụm tại trung vị của mỗi cụm
        indices = (labels == i)
        xtext, ytext = np.median(X[indices, :], axis=0)
        if not (np.isnan(xtext) or np.isnan(ytext)):
            txt = ax.text(xtext, ytext, str(i), fontsize=24)
            txt.set_path_effects([
                PathEffects.Stroke(linewidth=5, foreground="w"),
                PathEffects.Normal()])
            txts.append(txt)
    plt.title('t-sne visualization for thres={:.4f}'.format(thres))

gs = GridSpec(3, 4)
plt.figure(figsize = (25, 18))
plt.subplots_adjust(wspace=0.4,hspace=0.3)

for i, thres in enumerate(np.linspace(0.11, 0.14, 12)):
    dbscan = DBSCAN(eps=thres, min_samples=11, metric='euclidean')
    labels = dbscan.fit_predict(X_std)
    _plot_kmean_scatter(X_std, labels, gs[i], thres)
````

````python
print("\nTóm tắt DBSCAN theo eps:")
print("eps     | clusters | noise | noise% | max_size | min_size")
print("-"*60)
thresholds = np.linspace(0.11, 0.14, 12)
for eps in thresholds:
    db = DBSCAN(eps=eps, min_samples=11)
    lbl = db.fit_predict(X_std)
    n_cl = len(set(lbl)) - (1 if -1 in lbl else 0)
    n_no = np.sum(lbl == -1)
    sizes = [np.sum(lbl == k) for k in set(lbl) if k != -1]
    print(f"{eps:.4f} | {n_cl:8d} | {n_no:5d} | {100*n_no/len(lbl):5.1f}% | {max(sizes,default=0):8d} | {min(sizes,default=0):8d}")
````

## Hierarchical Clustering (phân cụm phân cấp)

**Hình minh họa dendrogram**
<p align="center">
  <img src="https://www.shutterstock.com/image-vector/dendrogram-diagram-representing-tree-hierarchical-600nw-2394698767.jpg" width="200" />
</p>

- Trục ngang (x): chỉ số (index) của các quan sát.  
- Trục dọc (y): giá trị thước đo khoảng cách (mức độ khác biệt) giữa các cụm.  
- Mỗi **node lá** (leaf node) đại diện cho một quan sát riêng lẻ.  
- Các **node cha** (non-leaf node) biểu thị cụm được hình thành sau khi hợp nhất hoặc trước khi phân chia.  
- Nếu vẽ một **đường ngang** cắt ngang toàn bộ dendrogram ở một độ cao nhất định, các nhánh bị cắt sẽ tương ứng với các **cụm không giao nhau** tại mức phân cấp đó.

Thuật toán phân cụm phân cấp có hai hướng tiếp cận chính, khác nhau về chiều thực hiện:

#### **1. Chiến lược hợp nhất (Agglomerative) – Bottom-up**

Đây là cách tiếp cận phổ biến nhất, đi **từ dưới lên trên**:

- Ban đầu: Mỗi điểm dữ liệu là một cụm độc lập (tương ứng các node lá).  
- Ở mỗi bước: Tìm và **hợp nhất** cặp cụm gần nhau nhất (theo thước đo khoảng cách đã chọn) thành một cụm mới.  
- Cụm mới này trở thành node cha ở mức cao hơn.  
- Tiếp tục lặp lại → số lượng cụm giảm dần cho đến khi chỉ còn một cụm duy nhất.  

Ưu điểm: Dễ triển khai, thường được dùng trong các thư viện như scikit-learn.

#### **2. Chiến lược phân chia (Divisive) – Top-down**

Ngược lại, cách này đi **từ trên xuống dưới**:

- Bắt đầu: Toàn bộ dữ liệu là một cụm lớn (node gốc).  
- Ở mỗi bước: Chọn một cụm hiện có và **phân chia** nó thành hai cụm con sao cho **mức độ tách biệt** giữa hai cụm con là lớn nhất (theo thước đo khoảng cách).  
- Lặp lại quá trình phân chia đệ quy cho đến khi mỗi quan sát thuộc một cụm riêng.  

Cả hai chiến lược đều tạo ra cùng một dạng **cây nhị phân** (binary tree):  
- Node gốc chứa toàn bộ dữ liệu.  
- Node lá là từng quan sát riêng lẻ.  
- Mỗi node quyết định (non-leaf) có đúng **hai node con**.  

Sự khác biệt chủ yếu nằm ở **hướng duyệt**: agglomerative đi từ lá lên gốc, divisive đi từ gốc xuống lá.
#### 2. Các phương pháp đo khoảng cách giữa hai cụm (Linkage methods)

| Linkage              | Công thức chính                                                                 | Đặc điểm                                                                                   | Nhạy cảm với outlier |
|----------------------|---------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------|----------------------|
| **Ward**             | $ d(S_1, S_2) = \frac{N_1 N_2}{N_1 + N_2}||\mathbf{m}_1 - \mathbf{m}_2\|^2 $ | Dựa trên giảm phương sai (variance reduction). Thường cho cụm hình cầu, cân bằng kích thước. | Ít                  |
| **Single** (nearest neighbor) | $ \min d(\mathbf{x}_i \in S_1, \mathbf{x}_j \in S_2) $                      | Dễ tạo cụm dạng chuỗi (chaining), nhạy cảm với nhiễu.                                       | Rất cao             |
| **Complete** (farthest neighbor) | $ \max d(\mathbf{x}_i \in S_1, \mathbf{x}_j \in S_2) $                       | Tạo cụm compact, hình cầu, nhưng phóng đại khoảng cách nếu có outlier.                      | Cao                 |
| **Average** (group average) | Trung bình tất cả cặp khoảng cách giữa hai cụm                                  | Cân bằng, ít bị ảnh hưởng bởi outlier hơn single/complete.                                  | Trung bình          |

- **Centroid linkage** cũng tồn tại nhưng ít dùng hơn.
- Với dữ liệu **non-Euclidean**, thay centroid bằng **clustroid** (điểm thực tế trong cụm có khoảng cách trung bình nhỏ nhất đến các điểm khác).

**Ward** và **average** thường được khuyến nghị trong thực tế khi dữ liệu có cấu trúc rõ ràng.

#### 3. Điều kiện dừng thuật toán

Vì thuật toán tạo toàn bộ dendrogram, bạn quyết định số cụm bằng cách:
- Cắt ngang dendrogram ở một mức khoảng cách nhất định.
- Khai báo trước số cụm mong muốn (n_clusters).
- Dựa trên **độ gắn kết** (cohesion): đường kính cụm, bán kính đến centroid, hoặc mật độ.

Các tiêu chí dựa trên đường kính/bán kính dễ bị ảnh hưởng bởi outlier.

#### 4. Độ phức tạp tính toán

- Mỗi bước tính khoảng cách giữa các cặp cụm → O(n²).
- Tổng cộng n-1 bước → O(n³) thời gian.
- Không gian lưu trữ cũng lớn → chỉ phù hợp dữ liệu nhỏ (vài nghìn điểm).

Có thể tối ưu xuống O(n² log n) bằng priority queue, nhưng vẫn hạn chế với dữ liệu lớn.

### Yếu tố quyết định “gần nhau nhất” – Linkage Criteria (rất quan trọng!)

Đây chính là phần làm nên sự khác biệt lớn trong kết quả:

| Linkage              | Ý nghĩa khoảng cách giữa 2 cụm A và B                                                                 | Đặc điểm hình dạng cụm                  | Nhạy cảm với nhiễu/outlier |
|----------------------|--------------------------------------------------------------------------------------------------------|------------------------------------------|-----------------------------|
| **Single** (nearest neighbor) | Khoảng cách nhỏ nhất giữa bất kỳ điểm nào trong A và B                                                | Dài, chuỗi, dạng "rắn"                   | Rất nhạy (chaining effect)  |
| **Complete** (farthest neighbor) | Khoảng cách lớn nhất giữa bất kỳ điểm nào trong A và B                                                | Cụm compact, hình cầu                    | Ít nhạy                     |
| **Average** (UPGMA)  | Trung bình khoảng cách tất cả cặp điểm giữa A và B                                                     | Cân bằng, ổn định                        | Trung bình                  |
| **Centroid**         | Khoảng cách giữa tâm (centroid) của A và B                                                             | Tốt khi cụm hình cầu                     | Nhạy với outlier            |
| **Ward**             | Tăng phương sai (variance) ít nhất khi gộp (tối thiểu hóa within-cluster variance)                    | Tốt nhất cho cụm hình cầu, cân bằng kích thước | Ít nhạy nhất                |

→ **Ward** và **Average** thường được khuyến nghị dùng mặc định trong nhiều bài toán thực tế.

### Ưu điểm cốt lõi của Hierarchical Clustering

- Không cần chỉ định trước số lượng cụm (dendrogram giúp chọn sau)
- Tạo ra cấu trúc phân cấp → hiểu được mối quan hệ giữa các cụm ở nhiều mức độ
- Dendrogram rất trực quan, dễ giải thích cho người không chuyên
- Linh hoạt với nhiều loại linkage → phù hợp nhiều hình dạng dữ liệu

### Nhược điểm chính

- Phức tạp tính toán: O(n²) đến O(n³) → chỉ phù hợp dữ liệu nhỏ/trung bình (< vài nghìn điểm)
- Một khi đã gộp/tách thì không thể sửa lại (greedy algorithm)
- Nhạy cảm với lựa chọn linkage và metric khoảng cách

# Anomaly Detection (Phát hiện bất thường)

**Anomaly Detection (Phát hiện bất thường)** là quá trình tự động xác định các điểm dữ liệu, sự kiện hoặc quan sát chệch hướng quá nhiều so với hành vi bình thường của toàn bộ tập dữ liệu. Các điểm này được gọi là *outliers* (ngoại lai), *anomalies* (bất thường) hoặc *exceptions* (ngoại lệ).

Trong thực tế, bất thường thường được chia làm 3 loại chính:

* **Bất thường điểm (Point Anomalies):** Một điểm dữ liệu đơn lẻ nằm cách xa toàn bộ phần còn lại.
  * *Ví dụ:* Một giao dịch thẻ tín dụng 1 tỷ VND giữa các giao dịch mua cà phê 50k.
* **Bất thường ngữ cảnh (Contextual Anomalies):** Một điểm dữ liệu có thể bình thường trong ngữ cảnh này nhưng lại bất thường trong ngữ cảnh khác.
  * *Ví dụ:* Nhiệt độ 35°C ở Việt Nam vào mùa hè là bình thường, nhưng ở Nam Cực là cực kỳ bất thường.
* **Bất thường tập hợp (Collective Anomalies):** Tập hợp các điểm dữ liệu liên tiếp cùng nhau tạo thành một mẫu bất thường, dù từng điểm tách rời có thể không bất thường.
  * *Ví dụ:* Hàng ngàn request truy cập web cùng một lúc trong cuộc tấn công DDoS.



**📌 Ứng dụng:** Phát hiện gian lận tài chính, chẩn đoán bệnh lý trong y tế, phát hiện lỗi hệ thống mạng, và làm sạch dữ liệu (Data Cleansing) trước khi huấn luyện các mô hình Machine Learning khác.

# Isolation Forest

Đa số các thuật toán truyền thống (như K-Means, DBSCAN, One-Class SVM) hoạt động theo nguyên lý: *"Cố gắng vẽ ra hình dáng của đám đông bình thường, ai lọt ra khỏi hình đó thì là bất thường"*. Tuy nhiên, việc vẽ chính xác "đám đông" trong không gian dữ liệu hàng trăm chiều là cực kỳ tốn thời gian và máy móc.

Thuật toán này đi ngược lại với tư duy thông thường là "đi tìm điểm bình thường" đó là "đi tìm điểm bất thường".

Thuật toán dựa trên 2 đặc tính cốt lõi của dữ liệu bất thường:

1. **Thiểu số (Few):** Số lượng điểm bất thường rất ít so với dữ liệu bình thường.
2. **Khác biệt (Different):** Giá trị của chúng thường nằm rất xa các điểm bình thường.



Vì "Ít" và "Khác biệt", các điểm bất thường rất dễ bị **cô lập (isolated)** ra khỏi phần còn lại của dữ liệu chỉ với một vài nhát cắt.

### Kỹ thuật lõi: Thuật toán hoạt động như thế nào?

Isolation Forest không dùng thước đo khoảng cách hay mật độ phức tạp, mà nó hoạt động dựa trên các cây quyết định gọi là **Isolation Trees (iTrees)**. Quá trình này giống như việc bạn nhắm mắt và "cắt bánh" ngẫu nhiên cho đến khi tách được từng miếng nhỏ chứa đúng 1 hạt nho.

**A. Quy trình xây dựng 1 cây iTree (Huấn luyện):**
Để "cô lập" các điểm dữ liệu, máy tính sẽ thực hiện các bước sau:
1. **Lấy mẫu:** Chọn ngẫu nhiên một tập con dữ liệu (subsample).
2. **Chọn trục cắt:** Chọn ngẫu nhiên một đặc trưng (feature/cột) bất kỳ.
3. **Cắt ngẫu nhiên:** Chọn ngẫu nhiên một giá trị nằm giữa mức Min và Max của đặc trưng đó, rồi vạch một đường thẳng cắt ngang không gian dữ liệu.
4. **Lặp lại:** Lặp lại việc cắt ngẫu nhiên này cho đến khi mỗi điểm dữ liệu bị nhốt vào một "phòng giam" riêng lẻ, hoặc cây đạt đến chiều cao giới hạn.

<img src=https://miro.medium.com/v2/resize:fit:720/format:webp/0*eNc-PZX1mFdqhF0V width="600" alt="Nhát cắt Isolation Forest">

*Minh họa trực quan: Điểm bất thường (đỏ) nằm chơ vơ ngoài rìa, chỉ cần 1-2 nhát cắt ngẫu nhiên là đã bị nhốt lại. Điểm bình thường (xanh) nằm chen chúc giữa đám đông, phải cắt chằng chịt hàng chục nhát mới tách riêng ra được.*

**B. Cấu trúc cây và Chiều dài đường đi (Path Length):**
Mỗi nhát cắt trên không gian 2D chính là một lần rẽ nhánh trên Cây quyết định.
* Số lượng nhát cắt cần thiết để cô lập một điểm $x$ được gọi là **chiều dài đường đi** $h(x)$.
* Vì một khu rừng có rất nhiều cây (iTrees), thuật toán sẽ tính giá trị trung bình của $h(x)$ trên tất cả các cây, ký hiệu là $E(h(x))$.
<img src=https://spotintelligence.com/wp-content/uploads/2024/05/illustration-isolation-forest.jpg width="600" alt="Nhát cắt Isolation Forest">

*Minh họa cấu trúc cây: Điểm bất thường có số tầng rẽ nhánh rất ngắn (nằm ngay gần gốc cây). Điểm bình thường có đường đi rất dài (nằm tít dưới ngọn lá).*

**C. Công thức chấm điểm (Anomaly Score):**
Dựa vào chiều dài đường đi trung bình, máy tính sẽ chấm điểm độ bất thường $s$ cho từng điểm dữ liệu dựa trên công thức toán học sau:

$$s(x, n) = 2^{-\frac{E(h(x))}{c(n)}}$$

Trong đó $c(n)$ là hằng số chuẩn hóa dựa trên kích thước mẫu $n$. Điểm số này sẽ hội tụ về khoảng từ 0 đến 1:

* **Nếu $s \approx 1$:** Điểm bị cô lập cực kỳ nhanh (đường đi ngắn) $\rightarrow$ **Chắc chắn là Bất thường (Anomaly)**.
* **Nếu $s \approx 0$:** Điểm chìm sâu trong đám đông (đường đi dài) $\rightarrow$ **Là điểm Bình thường (Normal)**.
* **Nếu $s \approx 0.5$:** Điểm nằm ở mức ranh giới, không quá rõ ràng.

**=> Kết luận phần kỹ thuật:** Isolation Forest tận dụng chính khuyết điểm của dữ liệu bất thường (ít và nằm xa) để "bắt" chúng một cách nhanh chóng, tiết kiệm tối đa tài nguyên tính toán so với việc đi đo lường hàng ngàn điểm dữ liệu bình thường!

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 1. TẠO DỮ LIỆU GIẢ LẬP
X_normal = 0.3 * np.random.randn(200, 2)
X_outliers = np.random.uniform(low=-4, high=4, size=(20, 2))
X = np.vstack((X_normal, X_outliers))

# 2. CHẠY ISOLATION FOREST
clf = IsolationForest(contamination=0.1, random_state=42)
labels = clf.fit_predict(X)

# 3. VẼ ĐƯỜNG VIỀN QUYẾT ĐỊNH (CONTOUR)
plt.figure(figsize=(10, 8))

xx, yy = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
# --- Chấm điểm độ bất thường (Anomaly Score) cho từng điểm trên lưới ---
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# --- Vẽ các vùng mật độ (Contourf) ---
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.Blues_r, alpha=0.3)

# --- Vẽ DUY NHẤT một đường ranh giới đỏ đứt nét (Decision Boundary) ---
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='red', linestyles='dashed')

# --- Vẽ lại các điểm dữ liệu lên trên bản đồ  ---
# Điểm bình thường (Label = 1)
plt.scatter(X[labels == 1][:, 0], X[labels == 1][:, 1], c='blue', label='Normal', s=40, zorder=3)
# Điểm bất thường (Label = -1)
plt.scatter(X[labels == -1][:, 0], X[labels == -1][:, 1], c='red', label='Anomaly', s=10, zorder=4)

plt.title("Isolation Forest Decision Boundary for Anomaly Detection", fontsize=14, fontweight='bold')
plt.xlabel(" X", fontsize=12)
plt.ylabel(" Y", fontsize=12)
plt.legend(loc='upper right')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
````

# Elliptic Envelope (Bao elip)

Khác với Isolation Forest dựa trên việc "cô lập" dữ liệu bằng các lát cắt ngẫu nhiên, **Elliptic Envelope** là một thuật toán dựa trên **thống kê (Statistical-based)**. Nó hoạt động dựa trên giả định rằng dữ liệu "bình thường" tuân theo một phân phối xác định và cố gắng vẽ một ranh giới bao quanh vùng tập trung nhất của dữ liệu đó.

### A. Nguyên lý cốt lõi: Giả định Phân phối Gaussian
Thuật toán này hoạt động hiệu quả nhất khi dữ liệu của bạn có dạng hình cụm và tuân theo **Phân phối chuẩn đa biến (Multivariate Gaussian Distribution)**.

* **Tư duy:** Thuật toán sẽ tính toán tâm (Mean) và độ phân tán (Covariance) của dữ liệu để tạo ra một hình elip (trong không gian 2D) hoặc một khối elipsoid (trong không gian nhiều chiều).
* **Điểm bất thường:** Bất kỳ điểm dữ liệu nào nằm ngoài "chiếc bao" elip này sẽ bị coi là outliers.



### B. Kỹ thuật lõi: Ước lượng Covariance mạnh mẽ (Fast-MCD)
Một vấn đề lớn của thống kê truyền thống là: nếu dữ liệu có sẵn Outliers, chúng sẽ làm lệch tâm và độ lệch chuẩn, khiến hình elip bị "kéo giãn" sai lệch. Để giải quyết, Elliptic Envelope sử dụng kỹ thuật **Minimum Covariance Determinant (MCD)**.

* **Chọn tập con (Pure Subset):** Thay vì tính toán trên toàn bộ dữ liệu, thuật toán tìm kiếm một tập con dữ liệu "sạch" nhất (thường chiếm $h\%$ tổng số điểm).
* **Cực tiểu hóa định thức (Minimizing Determinant):** Thuật toán tìm tập con sao cho định thức của ma trận hiệp phương sai (Covariance Matrix) là nhỏ nhất. Điều này tương đương với việc tìm vùng không gian có mật độ dữ liệu dày đặc nhất và ít bị phân tán nhất.
* **Khoảng cách Mahalanobis:** Để xác định một điểm $x$ nằm xa hay gần tâm elip, thuật toán không dùng khoảng cách Euclid thông thường mà dùng Khoảng cách Mahalanobis:
$$D_M(x) = \sqrt{(x - \mu)^T \Sigma^{-1} (x - \mu)}$$
    * Trong đó $\mu$ là tâm và $\Sigma$ là ma trận hiệp phương sai.
    * Khoảng cách này cực kỳ lợi hại vì nó tính đến cả hình dạng (độ co giãn) của cụm dữ liệu.



### C. Tham số quan trọng: Contamination (Mức độ nhiễm bẩn)
Đây là tham số quan trọng nhất mà người dùng cần cung cấp.

* Nó đại diện cho tỷ lệ phần trăm dữ liệu bất thường mà bạn dự đoán sẽ có trong tập dữ liệu (ví dụ: 0.05 tương đương 5%).
* Dựa vào tỷ lệ này, thuật toán sẽ quyết định "siết chặt" hay "mở rộng" vòng vây elip để loại bỏ đúng số lượng điểm ở vùng rìa có khoảng cách Mahalanobis lớn nhất.

## Bước 1: Phân Khúc Khách Hàng — Clustering với RFM Analysis

### Tại sao dùng RFM?

**RFM** là một framework phổ biến trong marketing và e-commerce, giúp định lượng hành vi khách hàng qua 3 chỉ số:

| Chỉ số | Viết tắt | Ý nghĩa | Cách tính |
|--------|----------|---------|-----------|
| **Recency** | R | Khách hàng mua lần cuối cách đây bao lâu? | Số ngày kể từ lần mua gần nhất |
| **Frequency** | F | Khách hàng mua bao nhiêu lần? | Số hóa đơn khác nhau |
| **Monetary** | M | Khách hàng chi bao nhiêu tiền? | Tổng giá trị giao dịch |

**Domain knowledge:** Khách hàng mua gần đây (R thấp), mua thường xuyên (F cao), và chi nhiều tiền (M cao) là khách hàng giá trị nhất. K-Means sẽ nhóm các khách hàng có đặc điểm RFM tương tự vào cùng một cluster.


````python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# ── Compute RFM features ────────────────────────────────────────────────────────
snapshot_date = retail_df["InvoiceDate"].max() + pd.Timedelta(days=1)

rfm = (
    retail_df.groupby("CustomerID")
    .agg(
        Recency=("InvoiceDate", lambda x: (snapshot_date - x.max()).days),
        Frequency=("InvoiceNo", "nunique"),
        Monetary=("TotalAmount", "sum"),
    )
    .reset_index()
)

print("RFM summary:")
print(rfm.describe().round(2))

# ── Scale features ──────────────────────────────────────────────────────────────
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[["Recency", "Frequency", "Monetary"]])

# ── Fit K-Means (k=4 chosen based on domain knowledge + elbow — see Part 2) ────
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)

# ── Cluster profile ─────────────────────────────────────────────────────────────
cluster_profile = rfm.groupby("Cluster")[["Recency", "Frequency", "Monetary"]].mean().round(1)
cluster_profile["Count"] = rfm.groupby("Cluster").size()
print("\nCluster Profile (mean values):")
print(cluster_profile.sort_values("Monetary", ascending=False))

# ── Visualization ───────────────────────────────────────────────────────────────
CLUSTER_LABELS = {0: "Champions", 1: "At Risk", 2: "New / Low-Value", 3: "Loyal Customers"}

# Auto-assign labels based on descending Monetary mean
sorted_clusters = cluster_profile["Monetary"].sort_values(ascending=False).index.tolist()
label_map = {c: list(CLUSTER_LABELS.values())[i] for i, c in enumerate(sorted_clusters)}
rfm["Segment"] = rfm["Cluster"].map(label_map)

colors = ["#e74c3c", "#3498db", "#95a5a6", "#2ecc71"]
palette = dict(zip(CLUSTER_LABELS.values(), colors))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter: Frequency vs Monetary (log scale for readability)
for seg, grp in rfm.groupby("Segment"):
    axes[0].scatter(grp["Frequency"], grp["Monetary"], label=seg,
                    alpha=0.5, s=18, color=palette.get(seg, "gray"))
axes[0].set_xlabel("Frequency (number of invoices)")
axes[0].set_ylabel("Monetary (total spend, GBP)")
axes[0].set_title("RFM Clustering: Frequency vs Monetary")
axes[0].legend(fontsize=8)

# Bar: average Monetary per segment
seg_means = rfm.groupby("Segment")["Monetary"].mean().reindex(list(CLUSTER_LABELS.values()))
seg_means.plot(kind="bar", ax=axes[1], color=colors, edgecolor="black")
axes[1].set_title("Average Monetary Value per Segment")
axes[1].set_xlabel("Segment")
axes[1].set_ylabel("Avg Total Spend (GBP)")
axes[1].tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.show()
````

### Diễn Giải Kết Quả RFM Clustering

Sau khi phân cụm, chúng ta gán nhãn kinh doanh cho từng nhóm dựa trên giá trị trung bình của R, F, M:

| Phân khúc | Recency | Frequency | Monetary | Hành động kinh doanh |
|-----------|---------|-----------|----------|---------------------|
|  **Champions** | Thấp | Cao | Cao | Reward, upsell, chương trình loyalty |
|  **Loyal Customers** | Trung bình | Khá cao | Khá cao | Gửi ưu đãi thành viên, giữ chân |
|  **At Risk** | Cao | Trung bình | Trung bình | Chiến dịch win-back, khảo sát churn |
|  **New / Low-Value** | Thấp–Cao | Thấp | Thấp | Onboarding, khuyến khích mua lần 2 |

> **Điểm quan trọng:** Clustering không tự đặt nhãn — đó là công việc của Domain Expert. Thuật toán chỉ nhóm các điểm dữ liệu tương tự nhau; ý nghĩa của từng nhóm phải được giải thích dựa trên kiến thức kinh doanh.


## Bước 2: Khám Phá Pattern — Association Rule Mining (Market Basket Analysis)

### Tại sao dùng Association Rules?

**Domain knowledge:** Trong e-commerce, hiểu được khách hàng hay mua sản phẩm nào cùng nhau cho phép doanh nghiệp:
- **Tăng doanh thu** qua cross-selling ("Sản phẩm thường mua kèm")
- **Tối ưu layout** website hoặc kệ hàng vật lý
- **Thiết kế bundle** khuyến mãi hiệu quả

Ba chỉ số quan trọng:

| Chỉ số | Công thức | Ý nghĩa |
|--------|-----------|---------|
| **Support** | P(A ∩ B) | Tỷ lệ đơn hàng chứa cả A và B |
| **Confidence** | P(B\|A) = P(A ∩ B) / P(A) | Khi mua A, xác suất cũng mua B |
| **Lift** | Confidence / P(B) | Lift > 1: A và B có liên hệ thực sự (không phải ngẫu nhiên) |

> **Lưu ý kỹ thuật:** Để tránh dữ liệu quá thưa (sparse), ta chỉ lấy **top 50 sản phẩm bán chạy nhất** — đây là một quyết định domain knowledge, không phải thuật toán.


````python
try:
    from mlxtend.frequent_patterns import apriori, association_rules
    from mlxtend.preprocessing import TransactionEncoder
except ImportError:
    import subprocess, sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "mlxtend", "-q"])
    from mlxtend.frequent_patterns import apriori, association_rules
    from mlxtend.preprocessing import TransactionEncoder

# ── Keep only the top-50 best-selling products (domain decision to reduce sparsity) ──
top_products = (
    retail_df.groupby("Description")["Quantity"]
    .sum()
    .nlargest(50)
    .index.tolist()
)

basket_df = retail_df[retail_df["Description"].isin(top_products)]

# ── Build invoice-product basket matrix (one-hot encoded) ──────────────────────
basket = (
    basket_df.groupby(["InvoiceNo", "Description"])["Quantity"]
    .sum()
    .unstack(fill_value=0)
    .reset_index(drop=True)
)
basket_bool = basket.applymap(lambda x: True if x > 0 else False)

print(f"Basket matrix shape: {basket_bool.shape}  (invoices × products)")

# ── Apriori — mine frequent itemsets ───────────────────────────────────────────
frequent_itemsets = apriori(basket_bool, min_support=0.02, use_colnames=True)
print(f"Frequent itemsets found: {len(frequent_itemsets)}")

# ── Generate association rules ─────────────────────────────────────────────────
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.5)
rules = rules.sort_values("lift", ascending=False).reset_index(drop=True)
print(f"Rules generated      : {len(rules)}")

# Sort by confidence and lift
rules_sorted = rules.sort_values(['confidence', 'lift'], ascending=[False, False])

if len(rules) > 0:
    print(f"\nTop 10 Association Rules:")
    print("=" * 80)

    # Display top rules in a readable format
    for i, (idx, rule) in enumerate(rules_sorted.head(10).iterrows(), 1):
        antecedents = ', '.join(list(rule['antecedents']))
        consequents = ', '.join(list(rule['consequents']))
        print(f"\nRule {i}:")
        print(f"  IF customer buys: {antecedents}")
        print(f"  THEN they will also buy: {consequents}")
        print(f"  Confidence: {rule['confidence']:.1%} | Lift: {rule['lift']:.2f} | Support: {rule['support']:.1%}")
else:
    print("\nNo rules found. Try lowering the confidence threshold or increasing the dataset size.")
````

### Diễn Giải Kết Quả Association Rules

Từ bảng top 10 rules, ta có thể đọc kết quả theo ngôn ngữ kinh doanh:

- **Antecedents → Consequents** với **Lift > 1.5** nghĩa là: "Khách hàng mua *[antecedents]* có khả năng cao hơn 1.5 lần so với ngẫu nhiên khi cũng mua *[consequents]*"
- **Confidence cao** → Quy tắc đáng tin cậy để đề xuất sản phẩm
- **Support cao** → Quy tắc áp dụng được cho nhiều đơn hàng (tầm ảnh hưởng lớn)

**Hành động kinh doanh từ kết quả:**
1. **Cross-sell:** Hiển thị gợi ý "Thường mua kèm" trên trang sản phẩm
2. **Bundle promotion:** Tạo combo giảm giá các sản phẩm hay đi cùng nhau
3. **Email marketing:** Gửi đề xuất cá nhân hóa dựa trên lịch sử mua hàng

> **Điểm quan trọng:** Support thấp không có nghĩa là rule vô nghĩa — một sản phẩm đặc biệt nhưng có Lift rất cao vẫn là thông tin quý giá cho các danh mục ngách (niche categories).


## Bước 3: Bảo Vệ Doanh Nghiệp — Anomaly Detection

> **Lưu ý:** Phần IV đã trình bày chi tiết các thuật toán Anomaly Detection (Isolation Forest, Local Outlier Factor, Elliptic Envelope). Tại đây, chúng ta tập trung vào **cách tích hợp** Anomaly Detection vào pipeline tổng thể và cách diễn giải kết quả theo góc độ kinh doanh.

### Tại sao cần Anomaly Detection trong bộ dữ liệu này?

**Domain knowledge cho phép ta xác định "bất thường":**
- **Hóa đơn có tổng tiền cực lớn** → Có thể là đơn hàng B2B hợp lệ, hoặc lỗi nhập liệu
- **Hóa đơn có số lượng sản phẩm cực cao nhưng tổng tiền thấp** → Có thể là thanh toán sai đơn vị (đặt số lượng = số tiền)
- **Hóa đơn có rất nhiều loại sản phẩm khác nhau** → Hành vi mua bất thường

Chúng ta sẽ xây dựng các đặc trưng ở **cấp độ hóa đơn** để phát hiện các bất thường này.


````python
from sklearn.ensemble import IsolationForest

# ── Build invoice-level features ───────────────────────────────────────────────
invoice_features = (
    retail_df.groupby("InvoiceNo")
    .agg(
        TotalAmount=("TotalAmount", "sum"),
        TotalQuantity=("Quantity", "sum"),
        UniqueProducts=("StockCode", "nunique"),
        AvgUnitPrice=("UnitPrice", "mean"),
    )
    .reset_index()
)

feature_cols = ["TotalAmount", "TotalQuantity", "UniqueProducts", "AvgUnitPrice"]
X = invoice_features[feature_cols]

# ── Isolation Forest — 1% contamination assumption ────────────────────────────
iso = IsolationForest(n_estimators=150, contamination=0.01, random_state=42)
iso.fit(X)                                                      # fit first
invoice_features["AnomalyScore"] = iso.decision_function(X)   # lower = more anomalous
invoice_features["IsAnomaly"] = iso.predict(X)                 # -1 = anomaly, 1 = normal

normal = invoice_features[invoice_features["IsAnomaly"] == 1]
anomaly = invoice_features[invoice_features["IsAnomaly"] == -1]

print(f"Normal invoices   : {len(normal):,}")
print(f"Anomalous invoices: {len(anomaly):,}  ({len(anomaly)/len(invoice_features)*100:.1f}%)")

print("\nSample anomalous invoices:")
from IPython.display import display
display(
    anomaly.sort_values("AnomalyScore").head(8)[
        ["InvoiceNo", "TotalAmount", "TotalQuantity", "UniqueProducts", "AvgUnitPrice"]
    ].reset_index(drop=True)
)
````

### Diễn Giải Kết Quả Anomaly Detection

Từ kết quả, các hóa đơn bị đánh dấu là bất thường thường có một trong các đặc điểm:

| Loại bất thường | Dấu hiệu | Hành động kinh doanh |
|----------------|----------|---------------------|
| **Đơn hàng cực lớn** | TotalAmount >> phân vị 99% | Review thủ công — có thể là B2B hoặc lỗi nhập liệu |

# Phần 2: Thách Thức Của Unsupervised Mining Trong Thực Tế

> Unsupervised Learning không có "đáp án đúng" để kiểm tra. Điều này tạo ra những thách thức đặc thù mà Supervised Learning không gặp phải.

## Thách Thức 1: Bài Toán "Tìm K"

Trong K-Means, chúng ta cần chọn số cluster K trước khi chạy thuật toán. Nhưng làm sao biết K nào là đúng khi **không có nhãn để kiểm tra**?

**Hai phương pháp phổ biến:**
1. **Elbow Method:** Vẽ đồ thị inertia (tổng bình phương khoảng cách đến centroid) theo K — tìm "cùi chỏ" nơi đồ thị bắt đầu phẳng dần
2. **Silhouette Score:** Đo mức độ một điểm dữ liệu "thuộc về" cluster của nó so với cluster khác. Giá trị từ -1 đến 1; **càng gần 1 càng tốt**

**Thực tế:** Hai phương pháp này chỉ là hướng dẫn — quyết định cuối cùng phải kết hợp với domain knowledge (ví dụ: "4 phân khúc khách hàng là đủ để marketing team quản lý").


````python
from sklearn.metrics import silhouette_score

# Re-use the scaled RFM data from Bước 1
k_range = range(2, 11)
inertias = []
silhouettes = []

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(rfm_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(rfm_scaled, labels, sample_size=2000, random_state=42))

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Elbow curve
axes[0].plot(list(k_range), inertias, "bo-", linewidth=2)
axes[0].axvline(x=4, color="red", linestyle="--", alpha=0.7, label="K=4 (chosen)")
axes[0].set_xlabel("Number of Clusters (K)")
axes[0].set_ylabel("Inertia (Within-cluster Sum of Squares)")
axes[0].set_title("Elbow Method")
axes[0].legend()

# Silhouette score
axes[1].plot(list(k_range), silhouettes, "rs-", linewidth=2)
axes[1].axvline(x=4, color="blue", linestyle="--", alpha=0.7, label="K=4 (chosen)")
best_k = k_range.start + int(np.argmax(silhouettes))
axes[1].axvline(x=best_k, color="green", linestyle=":", alpha=0.7, label=f"K={best_k} (peak silhouette)")
axes[1].set_xlabel("Number of Clusters (K)")
axes[1].set_ylabel("Silhouette Score")
axes[1].set_title("Silhouette Score")
axes[1].legend()

plt.suptitle("Choosing K: Elbow vs Silhouette — No Perfect Answer", fontsize=13, y=1.02)
plt.tight_layout()
plt.show()

print(f"Best K by silhouette: {best_k}")
print("Observation: The elbow and silhouette may suggest different K values.")
print("Final choice requires domain knowledge ('how many segments can our team manage?')")
````

## Thách Thức 2: Lời Nguyền Của Số Chiều (Curse of Dimensionality)

Khi số chiều (features) tăng lên, các điểm dữ liệu trong không gian nhiều chiều trở nên **cách đều nhau** — khoảng cách Euclidean mất đi ý nghĩa. Điều này làm hỏng các thuật toán dựa trên khoảng cách như K-Means, DBSCAN, KNN.

**Biểu hiện thực tế:**
- Silhouette score giảm khi thêm features nhiễu
- Clusters trở nên không rõ ràng và không ổn định
- Chi phí tính toán tăng theo hàm mũ

**Minh họa:** Ta thêm dần các chiều nhiễu (random noise) vào RFM và quan sát Silhouette Score giảm như thế nào.


````python
np.random.seed(42)
noise_dims = [0, 2, 5, 10, 20, 50, 100]
sil_scores = []

for n_noise in noise_dims:
    if n_noise == 0:
        X_noisy = rfm_scaled
    else:
        noise = np.random.randn(rfm_scaled.shape[0], n_noise)
        X_noisy = np.hstack([rfm_scaled, noise])

    km_temp = KMeans(n_clusters=4, random_state=42, n_init=10)
    labels_temp = km_temp.fit_predict(X_noisy)
    score = silhouette_score(X_noisy, labels_temp, sample_size=2000, random_state=42)
    sil_scores.append(score)
    print(f"  Dims: {3 + n_noise:>4}  |  Silhouette: {score:.4f}")

plt.figure(figsize=(8, 4))
plt.plot([3 + n for n in noise_dims], sil_scores, "go-", linewidth=2, markersize=7)
plt.axhline(y=sil_scores[0], color="red", linestyle="--", alpha=0.6, label=f"Baseline (3 dims): {sil_scores[0]:.3f}")
plt.xlabel("Total Number of Dimensions (3 RFM + noise)")
plt.ylabel("Silhouette Score")
plt.title("Curse of Dimensionality: Silhouette Score Degrades with Noise Dimensions")
plt.legend()
plt.tight_layout()
plt.show()
print("\nAs noise dimensions increase, cluster quality (silhouette) steadily degrades.")
print("Solution: Feature selection, PCA, or domain-knowledge-guided feature engineering.")
````

# Phần 3: Best Practices Cho Production

> Phần này trả lời câu hỏi: **"Doanh nghiệp thực sự tích hợp Unsupervised Mining vào hệ thống như thế nào?"**

## Quy Tắc 1: Scaling Là Bắt Buộc — Không Thể Bỏ Qua

K-Means sử dụng khoảng cách Euclidean. Nếu các features có đơn vị và thang đo khác nhau (ví dụ: Recency tính bằng ngày vs. Monetary tính bằng tiền), feature có giá trị lớn hơn sẽ **thống trị** kết quả phân cụm một cách không công bằng.

**Minh họa:** So sánh clustering trên dữ liệu RFM thô vs. đã chuẩn hóa.


````python
# Compare KMeans on raw RFM vs. scaled RFM
rfm_raw = rfm[["Recency", "Frequency", "Monetary"]].values

km_raw = KMeans(n_clusters=4, random_state=42, n_init=10)
labels_raw = km_raw.fit_predict(rfm_raw)

km_scaled = KMeans(n_clusters=4, random_state=42, n_init=10)
labels_scaled = km_scaled.fit_predict(rfm_scaled)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Raw (unscaled) — Monetary dominates
scatter0 = axes[0].scatter(rfm["Recency"], rfm["Monetary"],
                           c=labels_raw, cmap="tab10", alpha=0.4, s=15)
axes[0].set_xlabel("Recency (days)")
axes[0].set_ylabel("Monetary (GBP)")
axes[0].set_title("WITHOUT Scaling — Monetary Dominates\n(clusters split only on Monetary axis)")
axes[0].set_ylim(0, rfm["Monetary"].quantile(0.99))

# Scaled — fair clustering
scatter1 = axes[1].scatter(rfm["Recency"], rfm["Monetary"],
                           c=labels_scaled, cmap="tab10", alpha=0.4, s=15)
axes[1].set_xlabel("Recency (days)")
axes[1].set_ylabel("Monetary (GBP)")
axes[1].set_title("WITH StandardScaler — Balanced Clusters\n(all three RFM dimensions contribute equally)")
axes[1].set_ylim(0, rfm["Monetary"].quantile(0.99))

sil_raw = silhouette_score(rfm_raw, labels_raw, sample_size=2000, random_state=42)
sil_scaled = silhouette_score(rfm_scaled, labels_scaled, sample_size=2000, random_state=42)

print(f"Silhouette (raw data) : {sil_raw:.4f}")
print(f"Silhouette (scaled)   : {sil_scaled:.4f}")
print(f"Improvement           : +{(sil_scaled - sil_raw):.4f}")

plt.tight_layout()
plt.show()
````

## Quy Tắc 2: Giảm Số Chiều (Dimensionality Reduction) Trước Khi Cluster

Khi dataset có nhiều features, PCA (Principal Component Analysis) giúp:
1. **Giảm nhiễu** — loại bỏ các chiều có ít thông tin
2. **Trực quan hóa** — project xuống 2D để vẽ được cluster
3. **Tăng tốc độ** — ít chiều hơn = tính toán nhanh hơn

**Khi nào dùng PCA trước clustering?**
- Khi số features > ~10–20
- Khi có nhiều features tương quan với nhau (multicollinearity)
- Khi muốn visualize clusters trong 2D/3D

**Minh họa:** Áp dụng PCA trên dữ liệu RFM đã mở rộng (thêm các features từ invoice).


````python
from sklearn.decomposition import PCA

# Build an extended customer feature set (RFM + additional behavioral features)
customer_extended = (
    retail_df.groupby("CustomerID")
    .agg(
        Recency=("InvoiceDate", lambda x: (snapshot_date - x.max()).days),
        Frequency=("InvoiceNo", "nunique"),
        Monetary=("TotalAmount", "sum"),
        AvgOrderValue=("TotalAmount", lambda x: x.sum() / retail_df.loc[x.index, "InvoiceNo"].nunique()),
        UniqueProducts=("StockCode", "nunique"),
        AvgQuantityPerOrder=("Quantity", "mean"),
    )
    .reset_index()
)

feature_cols_ext = ["Recency", "Frequency", "Monetary",
                    "AvgOrderValue", "UniqueProducts", "AvgQuantityPerOrder"]

X_ext = StandardScaler().fit_transform(customer_extended[feature_cols_ext])

# ── PCA: explained variance ─────────────────────────────────────────────────────
pca_full = PCA(n_components=len(feature_cols_ext))
pca_full.fit(X_ext)

explained = pca_full.explained_variance_ratio_
cumulative = np.cumsum(explained)

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Scree plot
axes[0].bar(range(1, len(explained)+1), explained * 100, color="#3498db", edgecolor="black")
axes[0].plot(range(1, len(explained)+1), cumulative * 100, "ro-", linewidth=2, label="Cumulative")
axes[0].axhline(y=90, color="gray", linestyle="--", alpha=0.7, label="90% threshold")
axes[0].set_xlabel("Principal Component")
axes[0].set_ylabel("Variance Explained (%)")
axes[0].set_title("PCA Scree Plot — 6-Feature Customer Dataset")
axes[0].legend()

# 2D PCA projection colored by cluster (from scaled RFM above)
pca_2d = PCA(n_components=2, random_state=42)
# Use only customers present in both rfm and customer_extended
shared_ids = rfm.set_index("CustomerID").index.intersection(customer_extended["CustomerID"])
X_2d = pca_2d.fit_transform(
    StandardScaler().fit_transform(
        customer_extended[customer_extended["CustomerID"].isin(shared_ids)][feature_cols_ext]
    )
)
seg_colors = rfm[rfm["CustomerID"].isin(shared_ids)]["Segment"].values
unique_segs = list(dict.fromkeys(seg_colors))
color_map = dict(zip(unique_segs, ["#e74c3c", "#3498db", "#2ecc71", "#95a5a6"]))

for seg in unique_segs:
    mask = seg_colors == seg
    axes[1].scatter(X_2d[mask, 0], X_2d[mask, 1],
                    label=seg, alpha=0.4, s=15, color=color_map[seg])

axes[1].set_xlabel(f"PC1 ({explained[0]*100:.1f}% variance)")
axes[1].set_ylabel(f"PC2 ({explained[1]*100:.1f}% variance)")
axes[1].set_title("2D PCA Projection of 6-Feature Customer Data\n(colored by RFM cluster)")
axes[1].legend(fontsize=8)

plt.tight_layout()
plt.show()

print(f"With 2 PCs: {cumulative[1]*100:.1f}% of variance retained")
print(f"With 3 PCs: {cumulative[2]*100:.1f}% of variance retained")
````

## Pipeline Chuyên Nghiệp: Kết Hợp Tất Cả

### Trong Code — sklearn Pipeline

Thay vì chạy từng bước riêng lẻ (scale → PCA → cluster), một pipeline production-ready nên **đóng gói tất cả** vào một `sklearn.pipeline.Pipeline`. Lợi ích:
- **Không bị data leakage:** Scaler và PCA chỉ được fit trên training data, transform đúng cách trên new data
- **Dễ tái sử dụng:** Lưu pipeline bằng `joblib.dump()`, load lại sau
- **Dễ tuning:** Kết hợp với `GridSearchCV` để tìm hyperparameters tối ưu

### Trong Doanh Nghiệp — End-to-End ML Pipeline

Một hệ thống Unsupervised Mining thực tế trong doanh nghiệp thường gồm các bước:

```
[Data Sources]          [Preprocessing]        [Modeling]            [Action]
  Database       ──→   ETL / Cleaning   ──→   Clustering       ──→  CRM Segments
  Event Logs     ──→   Feature Eng.     ──→   Assoc. Rules     ──→  Recommender
  Transactions   ──→   Scaling / PCA    ──→   Anomaly Det.     ──→  Alert Queue
                              │
                        [Monitoring]
                    - Cluster drift detection
                    - Rule freshness check
                    - Anomaly threshold tuning
```

**Tần suất chạy lại:**
- **Clustering:** Hàng tuần hoặc hàng tháng (hành vi khách hàng thay đổi chậm)
- **Association Rules:** Hàng ngày hoặc theo mùa (trend thay đổi nhanh hơn)
- **Anomaly Detection:** Thời gian thực (real-time streaming) hoặc batch hàng ngày


````python
from sklearn.pipeline import Pipeline

# ── Build a reusable production pipeline ────────────────────────────────────────
clustering_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("pca",    PCA(n_components=2, random_state=42)),
    ("kmeans", KMeans(n_clusters=4, random_state=42, n_init=10)),
])

# Fit on the extended customer features
X_customers = customer_extended[feature_cols_ext].values
cluster_labels_pipeline = clustering_pipeline.fit_predict(X_customers)

print("Pipeline steps:")
for step_name, step_obj in clustering_pipeline.steps:
    print(f"  {step_name:10s}: {type(step_obj).__name__}")

print(f"\nPipeline fitted on {len(X_customers):,} customers")
print(f"Cluster distribution:")
unique, counts = np.unique(cluster_labels_pipeline, return_counts=True)
for u, c in zip(unique, counts):
    print(f"  Cluster {u}: {c:,} customers ({c/len(cluster_labels_pipeline)*100:.1f}%)")

# ── Visualize pipeline output in 2D PCA space ───────────────────────────────────
pca_transform = Pipeline([
    ("scaler", StandardScaler()),
    ("pca",    PCA(n_components=2, random_state=42)),
])
X_2d_pipeline = pca_transform.fit_transform(X_customers)

plt.figure(figsize=(8, 5))
scatter = plt.scatter(X_2d_pipeline[:, 0], X_2d_pipeline[:, 1],
                      c=cluster_labels_pipeline, cmap="tab10", alpha=0.5, s=20)
plt.colorbar(scatter, label="Cluster")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("Professional Pipeline Output: Clusters in 2D PCA Space\n"
          "(StandardScaler → PCA(2) → KMeans(4))")
plt.tight_layout()
plt.show()

# ── Save pipeline for future reuse ──────────────────────────────────────────────
import joblib
joblib.dump(clustering_pipeline, "customer_segmentation_pipeline.pkl")
print("\nPipeline saved as 'customer_segmentation_pipeline.pkl'")
print("To load: pipeline = joblib.load('customer_segmentation_pipeline.pkl')")
print("To predict on new data: pipeline.predict(new_customer_features)")
````

# Tóm Tắt Lý Thuyết — Unsupervised Mining

## Ba Trụ Cột Chính

### 1. Clustering (Phân Cụm)

**Mục tiêu:** Nhóm các điểm dữ liệu tương tự nhau vào cùng một cluster, mà không cần nhãn sẵn.

| Thuật toán | Ý tưởng | Ưu điểm | Nhược điểm |
|-----------|---------|---------|-----------|
| **K-Means** | Lặp lại: gán điểm vào centroid gần nhất → cập nhật centroid | Nhanh, đơn giản, dễ scale | Phải chọn K trước; nhạy cảm với outlier; chỉ tốt với cluster hình cầu |
| **DBSCAN** | Nhóm các điểm trong vùng mật độ cao | Tự tìm số cluster; xử lý được outlier | Khó chọn tham số ε và minPts; không tốt với mật độ không đều |
| **Hierarchical** | Xây dựng cây phân cấp (dendrogram) | Không cần chọn K trước; visualize bằng dendrogram | Chậm với dữ liệu lớn O(n²) |

**Đánh giá (không có nhãn):**
- **Silhouette Score** ∈ [-1, 1]: đo mức độ cluster-cohesion vs. cluster-separation
- **Davies-Bouldin Index**: càng thấp càng tốt
- **Inertia (WCSS)**: dùng để chọn K qua Elbow Method

---

### 2. Pattern Discovery — Association Rules (Luật Kết Hợp)

**Mục tiêu:** Tìm các mối quan hệ "thường xảy ra cùng nhau" trong các tập giao dịch.

**Công thức cốt lõi:**

$$\text{Support}(A \Rightarrow B) = \frac{|\text{transactions containing A and B}|}{|\text{all transactions}|}$$

$$\text{Confidence}(A \Rightarrow B) = \frac{\text{Support}(A \cup B)}{\text{Support}(A)}$$

$$\text{Lift}(A \Rightarrow B) = \frac{\text{Confidence}(A \Rightarrow B)}{P(B)}$$

| Chỉ số | Ngưỡng tốt | Ý nghĩa |
|--------|-----------|---------|
| Support | > 0.01 | Rule đủ phổ biến để có ý nghĩa thống kê |
| Confidence | > 0.5 | Rule đủ đáng tin cậy |
| **Lift** | **> 1** | Rule có giá trị thực sự (không phải ngẫu nhiên) |

**Thuật toán phổ biến:** Apriori, FP-Growth (nhanh hơn Apriori vì không sinh candidate)

---

### 3. Anomaly Detection (Phát Hiện Bất Thường)

**Mục tiêu:** Tìm các điểm dữ liệu khác biệt đáng kể so với phần còn lại.

| Phương pháp | Cơ chế | Phù hợp khi |
|------------|--------|------------|
| **Isolation Forest** | Cô lập điểm bất thường bằng random partitioning — cần ít lần chia hơn | Dữ liệu lớn, nhiều chiều, không cần giả định phân phối |
| **Local Outlier Factor (LOF)** | So sánh mật độ cục bộ của một điểm với các láng giềng | Anomaly nằm trong các vùng mật độ khác nhau |
| **Elliptic Envelope** | Giả định dữ liệu tuân theo phân phối Gaussian; tìm điểm nằm ngoài ellipse | Dữ liệu xấp xỉ Gaussian, ít chiều |

---

## Những Nguyên Tắc Vàng Khi Làm Unsupervised Mining

1. **Scaling là bắt buộc** trước K-Means, PCA, LOF (bất kỳ thuật toán dùng khoảng cách)
2. **Domain knowledge dẫn đường** — thuật toán tìm pattern, con người giải thích ý nghĩa
3. **Không có "đáp án đúng"** — validate bằng nhiều metrics + business intuition
4. **Iteration là bình thường** — thử nhiều K, nhiều features, nhiều thuật toán
5. **Pipeline hóa code** để đảm bảo reproducibility và tránh data leakage
